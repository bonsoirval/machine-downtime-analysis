```python
cat_dict = {v : len(df[df['category'] == v]) for k, v in enumerate(df['category'].unique())}
# select the highest 10 values
sorted_cat_dict = dict(sorted(cat_dict.items(), key=lambda x:x[1], reverse=True)[:9])
sorted_cat_dict

plt.barh(sorted_cat_dict.keys(), sorted_cat_dict.values())
plt.title("Paticipant Categories : Highest 10")
plt.ylabel('Categories')
plt.xlabel('Count [Category]')
plt.show()
```

```python
location_dict = {v:len(df[df['location'] == v]) for k, v in enumerate(df['location'].unique())}
location_dict = dict(sorted(location_dict.items(), key=lambda x:x[1], reverse=True)[:9])
plt.barh(location_dict.keys(), location_dict.values())
plt.ylabel('Participant Cities')
plt.xlabel('Count [Participants Cities]')
plt.title("Participant Cities")

plt.show()
```

1. What is the minimum and maximum amount asked for that got the deal

```python
done_deals = df[df['deal'] == True]

amount_asked_for = done_deals['askedFor']
min_asked_for = amount_asked_for.min()
max_asked_for = amount_asked_for.max()
```

```python
md(f"Minimum amount asked for : ${float(min_asked_for)}")
```

```python

md(f"Maximum amount asked for : ${float(max_asked_for)}")
```

2. Who won the different seasons?

```python
print("The winners of each season and episode are as shown below")
done_deals.groupby(['season', 'episode']).sum()['entrepreneurs'].str.split(',')
```

3. From what locations were the winners

```python
print("The winners of each season and episode are as shown below")
done_deals.groupby(['season', 'episode']).sum()[['entrepreneurs', 'location']].head()
```

4. On the average, how much funding are asked for by those who won?

```python
md(f"The average amound won by the winners is ${round(done_deals['askedFor'].mean(), 2)}")
```
